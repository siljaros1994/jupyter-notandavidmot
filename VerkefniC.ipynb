```python
import warnings
warnings.filterwarnings('ignore')

import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
import ipywidgets
import datetime
```

```python
file = open("veitingarstadur.jpeg", "rb")
image = file.read()

image_headline = widgets.Image(value=image,format='jpeg',width='300')

label_headline = widgets.Label(value=' ',style={'description_width': 'initial'})

vbox_headline = widgets.VBox([image_headline, label_headline])

name = widgets.Text(placeholder='Your name here')
#date = widgets.DatePicker(description='Date')

obj = ipywidgets.DatePicker( description='Pick a Date', disabled=False, value=datetime.datetime.now(),)
date = widgets.DatePicker(value=datetime.datetime(2022, 1, 14, 14, 41, 54, 26999), description='Pick a Date')


vinir = widgets.IntSlider(value=3,min=0, max=10,step=1,style={'description_width': 'initial', 'handle_color': '#16a085'} )

button_send = widgets.Button(description='Bóka',tooltip='Send',style={'description_width': 'initial'})

output = widgets.Output()

def on_button_clicked(event):
    with output:
        clear_output()
        print("Sent message: ")
        print(f"Velkominn á Fjallkonuna {name.value}.")
        print(f" Þú ert með bókað borð þann {date.value} ")
        print(f" Fyrir {vinir.value} ")

button_send.on_click(on_button_clicked)

vbox_result = widgets.VBox([button_send, output])


text_0 = widgets.HTML(value="<h1>Velkominn á Fjallkonuna</h1>")
text_1 = widgets.HTML(value="<h2> </h2>")
text_2= widgets.HTML(value="<h2>Hvenær má bjóða þér að koma ? </h2>")
text_3= widgets.HTML(value="<h2> </h2>")
text_4= widgets.HTML(value="<h2>Hversu margir verða með þér ? </h2>")
text_5= widgets.HTML(value="<h2>Endilega farðu yfir pöntunua áður en þú sendir bókununa.</h2>")

vbox_text = widgets.VBox([text_0, text_1, name, text_2, date, text_3, text_4, vinir, text_5, vbox_result])


page = widgets.HBox([vbox_headline, vbox_text])
display(page)



```
